```python
# Feature Scaling in Machine Learning

# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Loading the dataset
data = pd.read_csv("Social_Network_Ads.xls")
data.head()
```

What is Feature Scaling?

Feature scaling is a crucial preprocessing step in machine learning where numerical features are transformed to a common scale. It ensures that all features contribute equally to the model, preventing bias towards features with larger magnitudes.
Why is Feature Scaling Important?

    Improved Model Performance:
        Faster Convergence: Scaled features speed up the convergence of gradient-based algorithms like Logistic Regression.
        Better Accuracy: Prevents features with large ranges from dominating the training process.
        Enhanced Generalization: Helps models perform better on unseen data.

    Fair Feature Comparison:
        Ensures that all features contribute equally, avoiding bias in model decision-making.

    Efficient Gradient Descent:
        Prevents slow convergence or poor optimization when feature scales vary significantly.



Visualizing the Data

To understand the range and relationship between features, let's plot them:

```python
# Scatter plot to visualize Age vs. Estimated Salary
sns.scatterplot(data=data, x='Age', y='EstimatedSalary')
plt.title('Age vs Estimated Salary (Original Data)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```

# Common Feature Scaling Techniques
1. Min-Max Scaling (Normalization)

    Transforms features to a fixed range, typically [0, 1].
    Formula:
    xscaled=xmin(x)max(x)min(x)
    xscaled=max(x)min(x)xmin(x)



```python
# Applying Min-Max Scaling
scaler_minmax = MinMaxScaler()
data_scaled_minmax = data.copy()
data_scaled_minmax[['Age', 'EstimatedSalary']] = scaler_minmax.fit_transform(data[['Age', 'EstimatedSalary']])

# Scatter plot for scaled data
sns.scatterplot(data=data_scaled_minmax, x='Age', y='EstimatedSalary')
plt.title('Age vs Estimated Salary (Min-Max Scaled)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```

2. Standardization (Z-score Normalization)

    Transforms features to have a mean of 0 and a standard deviation of 1.
    Formula:
    xscaled=xmean(x)std(x)
    xscaled=std(x)xmean(x)



```python
# Applying Standardization
scaler_standard = StandardScaler()
data_scaled_std = data.copy()
data_scaled_std[['Age', 'EstimatedSalary']] = scaler_standard.fit_transform(data[['Age', 'EstimatedSalary']])

# Scatter plot for scaled data
sns.scatterplot(data=data_scaled_std, x='Age', y='EstimatedSalary')
plt.title('Age vs Estimated Salary (Standardized)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```
