### Matrix Multiplication and Dot Product

Matrix operations, particularly matrix multiplication, are foundational in many fields, including computer science, physics, and machine learning. This notebook will explore matrix multiplication, scalar multiplication, and element-wise operations, and demonstrate how to visualize and perform these operations using Python.

### Scalar vs Matrix Multiplication

Matrix operations can be divided into scalar multiplication and matrix multiplication. In scalar multiplication, each element of the matrix is multiplied by a scalar value, while matrix multiplication involves taking the dot product of rows and columns between two matrices.

    Scalar Multiplication: Every element of a matrix is multiplied by a constant scalar value.
    Matrix Multiplication: The dot product is taken between rows of the first matrix and columns of the second matrix.

### Scalar Multiplication

In scalar multiplication, every element of matrix AA is multiplied by a scalar value.

```python
import numpy as np
# Define matrix A
A = np.array([[1, 2], [3, 4]])
# Define scalar value
scalar = 2
# Perform scalar multiplication
E = scalar * A
print("Scalar Multiplication:\n", E)
```

### Matrix Multiplication (Dot Product)

Matrix multiplication is performed by calculating the dot product of rows of the first matrix with columns of the second matrix. The number of columns in the first matrix must match the number of rows in the second matrix.

```python
# Define matrix B
B = np.array([[5, 6], [7, 8]])
# Perform matrix multiplication (Dot product)
F = np.dot(A, B)
print("Matrix Multiplication:\n", F)
```

### Element-wise Operations (Hadamard Product)

The element-wise multiplication (also called the Hadamard product) involves multiplying corresponding elements from two matrices of the same size.



```python
# Element-wise multiplication
G = A * B
print("Element-wise Multiplication:\n", G)
```

### Visualizing Matrix Operations

Visualizing matrices and matrix operations helps understand how they work. We will visualize matrix AA, matrix BB, and the results of scalar and matrix multiplication.

```python
import matplotlib.pyplot as plt

# Visualize matrices and results of operations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Matrix A visualization
ax1.matshow(A, cmap='Blues')
ax1.set_title('Matrix A')
ax1.set_xticks(np.arange(len(A[0])))
ax1.set_yticks(np.arange(len(A)))
ax1.set_xlabel('Columns')
ax1.set_ylabel('Rows')

# Scalar multiplication visualization
ax2.matshow(E, cmap='Blues')
ax2.set_title('Scalar Multiplication (A * 2)')
ax2.set_xticks(np.arange(len(E[0])))
ax2.set_yticks(np.arange(len(E)))
ax2.set_xlabel('Columns')
ax2.set_ylabel('Rows')

plt.tight_layout()
plt.show()
```

### Matrix Multiplication Visualization

Let's visualize the process of matrix multiplication by showing the input matrices and the resulting matrix. We will also visualize the intermediate steps, like element-wise multiplication of rows and columns.

```python
# Perform matrix multiplication (if dimensions are valid)
if A.shape[1] != B.shape[0]:
    print("Error: Invalid matrix dimensions for multiplication.")
else:
    C = np.dot(A, B)

    # Visualize matrices A, B, and C
    fig, (ax3, ax4, ax5) = plt.subplots(1, 3, figsize=(15, 5))

    # Matrix A visualization
    ax3.matshow(A, cmap='Blues')
    ax3.set_title('Matrix A')
    ax3.set_xticks(np.arange(len(A[0])))
    ax3.set_yticks(np.arange(len(A)))
    ax3.set_xlabel('Columns (of A)')
    ax3.set_ylabel('Rows (of A)')

    # Matrix B visualization
    ax4.matshow(B, cmap='Greens')
    ax4.set_title('Matrix B')
    ax4.set_xticks(np.arange(len(B[0])))
    ax4.set_yticks(np.arange(len(B)))
    ax4.set_xlabel('Columns (of B)')
    ax4.set_ylabel('Rows (of B)')

    # Resulting matrix C visualization
    ax5.matshow(C, cmap='Reds')
    ax5.set_title('Matrix Multiplication (C = A * B)')
    ax5.set_xticks(np.arange(len(C[0])))
    ax5.set_yticks(np.arange(len(C)))
    ax5.set_xlabel('Columns (of C)')
    ax5.set_ylabel('Rows (of C)')

    plt.tight_layout()
    plt.show()
```

### Real-World Applications of Matrix Multiplication

Matrix multiplication plays a crucial role in various real-world applications:

    Machine Learning: In neural networks, matrix multiplication is used to compute the weighted sum of inputs at each layer.
    Computer Graphics: Transformation matrices are used to rotate, scale, and translate objects in 2D or 3D space.
    Physics: Matrix operations are used in simulations to solve linear systems and model physical phenomena.
    Economics: Matrices are used to represent and solve systems of equations, such as in input-output models for industries.
