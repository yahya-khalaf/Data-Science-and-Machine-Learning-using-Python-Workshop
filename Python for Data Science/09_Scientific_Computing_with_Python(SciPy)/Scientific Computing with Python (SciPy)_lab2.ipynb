### Scientific Computing with Python (SciPy)

SciPy is a powerful library built on top of NumPy, providing many algorithms and functions for scientific computing, 
data analysis, and engineering. It's commonly used in various scientific applications due to its rich set of features.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.optimize import minimize
from scipy.interpolate import interp1d
from scipy.signal import savgol_filter
from scipy import stats
```

### 1. Numerical Integration using SciPy

Let's define a function `f(x) = sin(x)` and integrate it numerically over the interval from 0 to Ï€/2.


```python
def f(x):
    return np.sin(x)

# Perform the integration using scipy.integrate.quad
result, error = quad(f, 0, np.pi/2)
print("Integration result:", result)
print("Integration error estimate:", error)

# Visualizing the function and the area under the curve
x = np.linspace(0, np.pi/2, 100)
y = f(x)

plt.plot(x, y)
plt.fill_between(x, y, color='gray', alpha=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Integration of sin(x)')
plt.grid(True)
plt.show()
```

### 2. Numerical Optimization

Now, let's minimize a simple quadratic function `f(x) = x^2 - 4x + 3` using SciPy's optimize module.


```python
def objective_function(x):
    return x**2 - 4*x + 3

# Perform the minimization
result = minimize(objective_function, x0=2)
print("The minimum value of the function is:", result.fun)
print("The value of x at the minimum is:", result.x)

# Visualize the function and the minimum point
x = np.linspace(-2, 6, 100)
y = objective_function(x)

plt.plot(x, y)
plt.scatter(result.x, result.fun, color='red')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Minimizing x^2 - 4x + 3')
plt.grid(True)
plt.show()
```

### 3. Solving Linear Equations

Consider the following system of linear equations:
- 2x + y = 8
- x - 3y = -3

We will solve it using NumPy's linear algebra module.


```python
A = np.array([[2, 1], [1, -3]])
b = np.array([8, -3])

# Solve the system using np.linalg.solve
x = np.linalg.solve(A, b)
print("Solution to the system of equations:", x)

# Visualizing the lines and solution point
x_values = np.linspace(-10, 10, 100)
y1_values = (8 - 2 * x_values) / 1
y2_values = (x_values + 3) / 3

plt.plot(x_values, y1_values, label='2x + y = 8')
plt.plot(x_values, y2_values, label='x - 3y = -3')
plt.scatter(x[0], x[1], color='red', label='Solution')
plt.xlabel('x')
plt.ylabel('y')
plt.title('System of Linear Equations')
plt.legend()
plt.grid(True)
plt.show()
```

### 4. Interpolation using SciPy

Create an interpolation function using cubic spline interpolation and visualize it.


```python
x = np.linspace(0, 10, 10)
y = np.sin(x) + np.random.randn(10) * 0.5  # Noisy data

# Create an interpolation function
f = interp1d(x, y, kind='cubic')

# Generate new x-values for interpolation
x_new = np.linspace(0, 10, 100)
y_new = f(x_new)

# Plot the original data and interpolated data
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x_new, y_new, label='Interpolated Data')
plt.legend()
plt.show()
```

### 5. Smoothing Data using Savitzky-Golay filter

Applying a Savitzky-Golay filter to smooth out noisy data.


```python
y_smooth = savgol_filter(y, window_length=51, polyorder=3)

# Plot the original and smoothed data
plt.plot(x, y, label='Original Data')
plt.plot(x, y_smooth, label='Smoothed Data')
plt.legend()
plt.show()
```

### 6. Hypothesis Testing: One-Sample t-test

Performing a one-sample t-test to check if the population mean is significantly different from a hypothesized value.


```python
data = np.array([10, 12, 15, 18, 20])
t_stat, p_value = stats.ttest_1samp(data, 15)

print("t-statistic:", t_stat)
print("p-value:", p_value)

# Interpretation of the results
if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

### 7. Correlation Analysis

Let's generate some correlated data and calculate the Pearson correlation coefficient.


```python
x = np.random.randn(100)
y = 2 * x + np.random.randn(100)

# Calculate the Pearson correlation coefficient
corr_coef, p_value = stats.pearsonr(x, y)

print("Pearson correlation coefficient:", corr_coef)
print("p-value:", p_value)
```

### 8. Data Visualization and Descriptive Statistics

Generating random data and calculating descriptive statistics like mean, median, and standard deviation.


```python
data = np.random.randn(100)
mean = np.mean(data)
std_dev = np.std(data)
median = np.median(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)
print("Median:", median)

# Visualizing the data with a histogram and box plot
plt.hist(data, bins=20)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data")
plt.show()

plt.boxplot(data)
plt.title("Box Plot of Random Data")
plt.show()
```
