### 2D Vectors

You can create and manipulate 2D vectors. Here's an example with vectors uu and vv.

```python
# 2D vectors
# create simple 2d vectors
u = np.array([2, 5])
v = np.array([3, 1])
u,v
```

### Plotting 2D Vectors

We can plot 2D vectors as points on a graph using Matplotlib.
Code Cell:

```python
# They can be represented graphically on a 2D graph, as points:
import matplotlib.pyplot as plt
x_coords, y_coords = zip(u, v)  # Extract x and y coordinates from vectors u and v
# Create a scatter plot using matplotlib
plt.scatter(x_coords, y_coords, color=["r", "b"])  # Plot points with red and blue colors
# Set axis limits for the plot
plt.axis([0, 9, 0, 6])  # Set x-axis from 0 to 9, y-axis from 0 to 6
# Enable grid lines on the plot
plt.grid()  # Display grid lines for better readability
# Display the plot
plt.show()  # Make the plot visible
```

### Plotting 2D Vectors as Arrows

We can create a small function to plot 2D vectors as arrows.


```python
# Vectors can also be represented as arrows. Let's create a small convenience function to draw arrows:
def plot_vector2d(vector2d, origin=[0, 0], **options): 
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],#plot 2D vectors as an arrow on a graph
              head_width=0.2, head_length=0.3, length_includes_head=True,
              **options)
```

```python
plot_vector2d(u, color="r") #Plot Vector u with red color
plot_vector2d(v, color="b") #Plot Vector v with blue color
plt.axis([0, 9, 0, 6])  #Set x-axis from 0 to 9, y-axis from 0 to 6
plt.grid()
plt.show()
```

### 3D Vectors

```python
# 3D vectors
# Plotting 3D vectors is also relatively straightforward. First let's create two 3D vectors:
a = np.array([1, 2, 8])
b = np.array([5, 6, 3])
a,b
```

### Plotting 3D Vectors

We can plot 3D vectors using Axes3D in Matplotlib.


```python
# plot them using matplotlib's Axes3D:
# Create a subplot as a 3D projection 
subplot3d = plt.subplot(111, projection='3d') # 111 for single subplot, '3d' for projection
# Extract x, y, and z coordinates from the vectors
x_coords, y_coords, z_coords = zip(a,b)
# Plot the vectors as 3D scatter points
subplot3d.scatter(x_coords, y_coords, z_coords)
# Set limits for the z-axis
subplot3d.set_zlim3d([0, 9])
plt.show()
```

### Adding Vertical Lines to 3D Vectors

We can add vertical lines to the 3D vectors for better visualization.

```python
#Add vertical lines. Creating a small convenience function to plot a list of 3d vectors with vertical lines attached:
def plot_vectors3d(ax, vectors3d, z0, **options):
    for v in vectors3d:    # Loop through each vector in the list
        x, y, z = v  # Unpack the x, y, and z coordinates of the current vector
        # Plot a vertical line from the vector's position (x, y) to the base plane (z=z0)
        ax.plot([x,x], [y,y], [z0, z], color="gray", linestyle='dotted', marker=".")
        # Extract x, y, and z coordinates from all vectors for the scatter plot    
    x_coords, y_coords, z_coords = zip(*vectors3d)
        # Plot the vectors as points on the 3D axes using scatter plot with optional customizations
    ax.scatter(x_coords, y_coords, z_coords, **options)
# Create a 3D subplot using matplotlib
subplot3d = plt.subplot(111, projection='3d')
subplot3d.set_zlim([0, 9])
plot_vectors3d(subplot3d, [a,b], 0, color=("r","b"))
plt.show()
```

### Vector Norm

The norm (or length) of a vector can be calculated using Python or NumPy.

```python
# Norm
# We could implement this easily in pure python.
def vector_norm(vector):
    squares = [element**2 for element in vector]
    return sum(squares)**0.5

print("||", u, "|| =")
vector_norm(u)
```

```python
# However, it is much more efficient to use NumPy's norm function, available in the linalg (Linear Algebra) module:
import numpy.linalg as LA
LA.norm(u)
```

### Visualizing Vector Length

### Geometric Translation

By adding a vector vv to a shape defined by points, we can shift the entire shape.


```python
# f you have a shape defined by a number of points (vectors), and you add a vector v to all of these points, 
# then the whole shape gets shifted by v. 
# This is called a geometric translation:
t1 = np.array([2, 0.25])
t2 = np.array([2.5, 3.5])
t3 = np.array([1, 2])

x_coords, y_coords = zip(t1, t2, t3, t1)
plt.plot(x_coords, y_coords, "c--", x_coords, y_coords, "co")

plot_vector2d(v, t1, color="r", linestyle=":")
plot_vector2d(v, t2, color="r", linestyle=":")
plot_vector2d(v, t3, color="r", linestyle=":")

t1b = t1 + v
t2b = t2 + v
t3b = t3 + v

x_coords_b, y_coords_b = zip(t1b, t2b, t3b, t1b)
plt.plot(x_coords_b, y_coords_b, "b-", x_coords_b, y_coords_b, "bo")

plt.text(4, 4.2, "v", color="r", fontsize=18)
plt.text(3, 2.3, "v", color="r", fontsize=18)
plt.text(3.5, 0.4, "v", color="r", fontsize=18)

plt.axis([0, 6, 0, 5])
plt.grid()
plt.show()
```

### Dot product
    The dot product (also called scalar product or inner product in the context of the Euclidian space) 
    of two vectors u and v is a useful operation that comes up fairly often in linear algebra.
    In python
    The dot product is pretty simple to implement:


```python
def dot_product(v1, v2):
    return sum(v1i * v2i for v1i, v2i in zip(v1, v2))

dot_product(u, v)
```

```python
# But a much more efficient implementation is provided by NumPy with the dot function:
np.dot(u,v)
```

```python
# Equivalently, you can use the dot method of ndarrays:
u.dot(v)
```

```python
# Caution: the * operator will perform an elementwise multiplication, NOT a dot product:
print("  ",u)
print("* ",v, "(NOT a dot product)")
print("-"*10)

u * v
```

### Angle between vectors

```python
# Calculating the angle between vectors
# Let's use this formula to calculate the angle between u and v (in radians):
def vector_angle(u, v):
    cos_theta = u.dot(v) / LA.norm(u) / LA.norm(v)
    return np.arccos(np.clip(cos_theta, -1, 1))

theta = vector_angle(u, v)
print("Angle =", theta, "radians")
print("      =", theta * 180 / np.pi, "degrees")
```

### Vector Projection

```python
# Projecting a point onto an axis
# The dot product is also very useful to project points onto an axis.
u_normalized = u / LA.norm(u)
proj = v.dot(u_normalized) * u_normalized

plot_vector2d(u, color="r")
plot_vector2d(v, color="b")

plot_vector2d(proj, color="k", linestyle=":")
plt.plot(proj[0], proj[1], "ko")

plt.plot([proj[0], v[0]], [proj[1], v[1]], "b:")

plt.text(1, 2, "$proj_u v$", color="k", fontsize=18)
plt.text(1.8, 0.2, "$v$", color="b", fontsize=18)
plt.text(0.8, 3, "$u$", color="r", fontsize=18)

plt.axis([0, 8, 0, 5.5])
plt.grid()
plt.show()
```


#### Some vocabulary
    In NumPy, each dimension is called an axis.
    The number of axes is called the rank.
        For example, the above 3x4 matrix is an array of rank 2 (it is 2-dimensional).
        The first axis has length 3, the second has length 4.
    An array's list of axis lengths is called the shape of the array.
        For example, the above matrix's shape is (3, 4).
        The rank is equal to the shape's length.
    The size of an array is the total number of elements, which is the product of all axis lengths (eg. 3*4=12)

### Reshaping an array
    Changing the shape of an ndarray is as simple as setting its shape attribute. 
    However, the array's size must remain the same.


```python
# In place
g = np.arange(24)
print(g)
print("Rank:", g.ndim)
```

```python
g.shape = (6, 4)
print(g)
print("Rank:", g.ndim)
```

```python
g.shape = (2, 3, 4)
print(g)
print("Rank:", g.ndim)
```

```python
# reshape
# The reshape function returns a new ndarray object pointing at the same data.
# This means that modifying one array will also modify the other.
g2 = g.reshape(4,6)
print(g2)
print("Rank:", g2.ndim)
```

```python
# Set item at row 1, col 2 to 999 (more about indexing below).
g2[1, 2] = 999
g2
 
```
