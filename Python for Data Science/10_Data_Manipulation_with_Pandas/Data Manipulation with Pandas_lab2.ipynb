### Data Manipulation with Python

Data manipulation refers to the process of transforming data to extract insights, prepare it for analysis, or perform calculations. This involves operations such as filtering, sorting, aggregating, merging, and transforming data. In this notebook, weâ€™ll work with both JSON and structured datasets to explore advanced manipulation techniques using Python libraries such as Pandas and NumPy.


```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

with open('North_Sea_Oil_Field.json', 'r') as f:
    data = json.load(f)

# Create a Pandas DataFrame from the JSON data
df = pd.DataFrame(data)
df.head()  # Display the first few rows of the dataset
```

### Extracting Columns as NumPy Arrays

You can extract specific columns as NumPy arrays for more advanced numerical operations.


```python
# Extract specific columns as arrays
well_ids = df['well_id'].values
latitudes = df['latitude'].values
depth = df['depth'].values
df
```

## Array Manipulation with NumPy
### A. Array Concatenation

Array concatenation is often used in data preparation to combine multiple arrays into one for downstream tasks.

```python
conc = np.concatenate([well_ids, latitudes, depth])
print("Concatenated Array:", conc)
```

### B.Array Splitting

You can split arrays into smaller chunks based on predefined positions. This is useful when segmenting data for analysis.


```python
x1, x2, x3 = np.split(depth, [5, 13])
x1, x2, x3
```

### Sorting and Filtering Data

You can sort your DataFrame by specific columns or filter rows based on conditions. Let's sort the data by latitude.

```python
df.sort_values(by='latitude', ascending=True, inplace=False)
```

### Dropping and Modifying Columns

We can drop unnecessary columns or modify values within the dataset. For example, let's remove the `longitude` column and replace specific values in the `depth` column.


```python
# Drop a column
df.drop('longitude', axis='columns', inplace=True)

# Replace values
df.replace(1150, 1200, inplace=True)
df.head()
```

### Renaming Columns

Renaming columns can improve readability or adapt the dataset for specific tasks.


```python
df.rename(columns={'production_rate': 'rate', 'age': 'age1'}, inplace=True)
df.head()
```

### Categorizing Continuous Variables

Binning continuous variables involves dividing them into intervals or categories. This is useful for summarizing data or preparing it for machine learning algorithms.


```python
# Define bins
bins = [5, 7, 10, 13, 16]

# Categorize (bin) the 'age1' column
cats = pd.cut(df['age1'], bins)
cats.value_counts()
```

### Custom Bin Labels

Assigning unique labels to bins can make the data more interpretable.


```python
bin_names = ['old', 'med_Age', 'young', 'new_born']
new_cats = pd.cut(df['age1'], bins, labels=bin_names)
pd.Series(new_cats).value_counts()
```

### Cumulative Sum

The cumulative sum of observations in each bin provides insights into data distribution.


```python
pd.Series(new_cats).value_counts().cumsum()
```
