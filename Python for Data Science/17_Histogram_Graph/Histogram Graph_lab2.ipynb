## Histogram Graphs

A histogram is a powerful tool for visualizing the distribution of numerical data and identifying patterns or outliers. This notebook covers:
- Basic histogram
- Custom bin sizes
- Normalized histograms
- Overlaid histograms
- Histograms with KDE (Kernel Density Estimation)

### Dataset Overview
The dataset contains information about universities worldwide, including rankings, scores for various criteria, and demographic statistics.


```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("timesData.csv")

# Preprocess the dataset
# Convert numerical columns to numeric types, handling missing values
data['teaching'] = pd.to_numeric(data['teaching'], errors='coerce')
data['international'] = pd.to_numeric(data['international'], errors='coerce')
data['research'] = pd.to_numeric(data['research'], errors='coerce')
data['citations'] = pd.to_numeric(data['citations'], errors='coerce')
data['income'] = pd.to_numeric(data['income'], errors='coerce')
data['total_score'] = pd.to_numeric(data['total_score'], errors='coerce')

# Display basic information
data.info()
```

### Basic Histogram
We start with a basic histogram to visualize the distribution of `teaching` scores.


```python
# Basic Histogram
plt.figure(figsize=(10, 6))
plt.hist(data['teaching'].dropna(), bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title("Distribution of Teaching Scores", fontsize=14)
plt.xlabel("Teaching Score", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y')
plt.show()
```

### Custom Bin Sizes
Adjusting the number of bins can reveal more granular or broader patterns in the data.


```python
# Custom Bin Sizes
plt.figure(figsize=(10, 6))
plt.hist(data['teaching'].dropna(), bins=50, color='green', alpha=0.7, edgecolor='black')
plt.title("Distribution of Teaching Scores with 50 Bins", fontsize=14)
plt.xlabel("Teaching Score", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y')
plt.show()
```

### Normalized Histogram
A normalized histogram displays relative frequencies, making it easier to compare datasets with different sizes.


```python
# Normalized Histogram
plt.figure(figsize=(10, 6))
plt.hist(data['teaching'].dropna(), bins=20, color='orange', alpha=0.7, density=True, edgecolor='black')
plt.title("Normalized Distribution of Teaching Scores", fontsize=14)
plt.xlabel("Teaching Score", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(axis='y')
plt.show()
```

### Overlaid Histograms
Overlaid histograms allow us to compare the distributions of multiple variables.


```python
# Overlaid Histograms
plt.figure(figsize=(12, 6))
plt.hist(data['teaching'].dropna(), bins=20, alpha=0.5, label="Teaching", color='blue', edgecolor='black')
plt.hist(data['research'].dropna(), bins=20, alpha=0.5, label="Research", color='red', edgecolor='black')
plt.title("Comparison of Teaching and Research Scores", fontsize=14)
plt.xlabel("Score", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.grid(axis='y')
plt.show()
```

### Histogram with KDE (Kernel Density Estimation)
Combining a histogram with a KDE plot provides a smoothed estimate of the data's distribution.


```python
# Histogram with KDE
plt.figure(figsize=(10, 6))
sns.histplot(data['citations'].dropna(), kde=True, color='purple', bins=20, alpha=0.7)
plt.title("Histogram with KDE for Citation Scores", fontsize=14)
plt.xlabel("Citation Score", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(axis='y')
plt.show()
```
