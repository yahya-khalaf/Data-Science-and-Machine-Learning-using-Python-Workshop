## Visualization Techniques - Line, Bar, and Scatter Plots

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "all_fuels_data.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

### Line Plot: Visualizing Trends Over Time
#### Description:

A line plot is used to visualize trends or changes over time. It's particularly useful for understanding how variables evolve chronologically.
#### Use Case: 
Ideal for tracking trends and changes over time.

```python
# Convert the 'date' column to datetime for better handling
data['date'] = pd.to_datetime(data['date'])

# Filter data for a specific commodity (e.g., Crude Oil)
crude_oil_data = data[data['commodity'] == 'Crude Oil']

# Plot a line graph of closing prices over time
plt.figure(figsize=(12, 6))
plt.plot(crude_oil_data['date'], crude_oil_data['close'], label='Closing Price', color='blue')

# Add title and labels
plt.title('Crude Oil Closing Prices Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Closing Price (USD)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

### Bar Plot: Comparing Trading Volumes
#### Description:

A bar plot is effective for comparing discrete categories or summarizing data over intervals. Here, we visualize the average trading volume for each year.
#### Use Case:
Best for comparing categorical data or summarized values.

```python
# Extract the year from the date
data['date'] = pd.to_datetime(data['date'])  # Ensure 'date' is in datetime format
data['year'] = data['date'].dt.year  # Add 'year' column to the main DataFrame

# Filter data for a specific commodity (e.g., Crude Oil)
crude_oil_data = data[data['commodity'] == 'Crude Oil']

# Calculate average trading volume per year
average_volume = crude_oil_data.groupby('year')['volume'].mean().reset_index()

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='year', y='volume', data=average_volume, palette='viridis')

# Add title and labels
plt.title('Average Trading Volume of Crude Oil Per Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Volume', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

### Scatter Plot: Relationship Between Price and Volume
#### Description:

A scatter plot shows the relationship between two variables. Its great for identifying correlations, trends, or clusters.
#### Use Case: 
Shows the relationship between two continuous variables.

```python
# Scatter plot of closing price vs. volume
plt.figure(figsize=(12, 6))
plt.scatter(crude_oil_data['close'], crude_oil_data['volume'], alpha=0.7, c=crude_oil_data['volume'], cmap='coolwarm')

# Add title and labels
plt.title('Relationship Between Crude Oil Price and Volume', fontsize=14)
plt.xlabel('Closing Price (USD)', fontsize=12)
plt.ylabel('Trading Volume', fontsize=12)
plt.colorbar(label='Volume Intensity')
plt.grid(True)
plt.show()
```
